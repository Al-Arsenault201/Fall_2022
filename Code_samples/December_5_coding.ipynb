IN the next step, we're going to read in review_1.txt and set up the tfidf matrix

```python
f = "/Users/alfredarsenault/Documents/GitHub/Fall_2022/Homeworks_labs_etc/Project_3_files/review_1.txt"
with open(f, "r") as infile:
    data = infile.read()
    word_list = data.split()
    print(word_list)
    print("there are ", len(word_list), " words in this review")
```

```python
tfidf = []
tfidf.append([word_list[0],1])
print(tfidf)
```

```python
for word in word_list[1:]:
    found = False
    for j in range(len(tfidf)):
        if word == tfidf[j][0]:   #  find out if this new word matches any existing word
            tfidf[j][1] += 1     # increment the number of times this word has been seen
            found == True
    if not found:
        # this is the first time we've seen this word; add a new row to the list
        tfidf.append([word, 1])
for i in range(len(tfidf)):
    print(tfidf[i])
```

## what's going on here
TF stands for term frequency. It's not the raw number of times a term appears in a review; it's what proportion of the total words in the review are that word.
To calculate TF we'll divide the raw numbers by the number of words; that is, len(word_list)

```python
for i in range(len(tfidf)):
    tfidf[i][1] /= len(word_list)
    print(tfidf[i])
    
```

### what's going on here:
- Now, we're going to read in review_2.txt. 
- We're going to split it into a list
- We're going to calculate the number of words in the review
- Then, we're going to add a new column to tfidf matrix for this review
- Initially, the new column will contain all 0's because we haven't encountered any words

```python
f2 = "/Users/alfredarsenault/Documents/GitHub/Fall_2022/Homeworks_labs_etc/Project_3_files/review_2.txt"
with open(f2, "r") as infile:
    data = infile.read()
    word_list = data.split()
    for k in range(len(tfidf)):
        tfidf[k].append(0)
        print(tfidf[k])
```

### what's next 
- read each word in review_2 (now contained in word_list)
- if it's already in tfidf - that is, it matches tfidf[z][0] for some value of z - increment
the current column by one
- otherwise create a new row; put the word in column 0; put a 0 in column 1 because we did not see this word in review_1, and put a  1 in column 2 because we have seen this word once in review_2

```python
for word in word_list: 
    found = False
    for z in range(len(tfidf)):
        if word == tfidf[z][0]:   #  find out if this new word matches any existing word
            tfidf[z][2] += 1     # increment the number of times this word has been seen
            found == True
    if not found:
        # this is the first time we've seen this word; add a new row to the list
        tfidf.append([word, 0,1])
for i in range(len(tfidf)):
    tfidf[i][2] /= len(word_list)
    print(tfidf[i])
    
```
