```python
#now cast everything to lower case to make future comparisons easier
for i in range(len(concatenated_data_list)): 
    for j in range(len(concatenated_data_list[i])):
        concatenated_data_list[i][j] = concatenated_data_list[i][j].lower()
#    print(concatenated_data_list[i])
    
```

```python
tfidf = []
word_list = concatenated_data_list[0]
tfidf.append([word_list[0],1])
for word in word_list[1:]:
    found = False
    for j in range(len(tfidf)):
        if word == tfidf[j][0]:   #  find out if this new word matches any existing word
            tfidf[j][1] += 1     # increment the number of times this word has been seen
            found = True
    if not found:
        # this is the first time we've seen this word; add a new row to the list
        tfidf.append([word, 1])
for i in range(len(tfidf)):
    print(tfidf[i])
```

# what's happening here
Now, we're going to write a loop that goes through the rest of the reviews
One review at a time
- We will add a column to tfidf list for each new review
- We will start by setting that column to 0 for every word so far
- Next we'll go through this new review one word at a time
- If we've see the word before, we increment the new column value by 1
- If not:
- add a new row, initiall with a list containing just this word
- add a 0 in the column for each review read so far
- then add a 1 in the column for this new review

```python
for r in range(1,len(concatenated_data_list)):  # read all the rest of the reviews
    
    for l in range(len(tfidf)):       # adds the new column with all entries = 0
        tfidf[l].append(0)
        
        
    word_list = concatenated_data_list[r]  # being lazy & choosing a shorter variable name
    for word in word_list:                 # for ever word in this new review
        found = False
        for z in range(len(tfidf)):
            if word == tfidf[z][0]:   #  find out if this new word matches any existing word
                tfidf[z][r+1] += 1     # increment the number of times this word has been seen
                found = True
        if not found:
        # this is the first time we've seen this word; add a new row to the list
            tfidf.append([word])
            for j in range(r):
                tfidf[-1].append(0)
            tfidf[-1].append(1)
for i in range(len(tfidf)):
    print(tfidf[i])
```

## what's happening next
now caculate tf by dividing each raw number of occurrences of a word by the frequency

```python
lengths = []
for i in range(len(concatenated_data_list)):
    lengths.append(len(concatenated_data_list[i]))
    print (lengths[i])
```

## Next
divide each element in tfidf matrix by the number of words in that review

```python
for i in range(len(tfidf)):
    for j in range(1, 10):
        tfidf[i][j] /= lengths[j-1]
    print(tfidf[i])
```

Now we have the tf matrix, and we have to calculate the idf matrix

```python
import math
#tfidf is the data so far
# we're going to create a new list, idf, that we'll calculate
idf = []
for i in range(len(tfidf)):    # for each row
    non_zero_cells = 0
    for j in range(1, len(tfidf[i])):
        if tfidf[i][j] > 0:
            non_zero_cells += 1
    idf.append(math.log(10/non_zero_cells))
 #       idf.append([tfidf[i][0], math.log(10/non_zero_cells)])
print (idf)
```
